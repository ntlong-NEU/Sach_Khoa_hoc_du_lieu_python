```python
# Đường dẫn file Excel
file_path = root_path + "data/VN_Economic_Indicators_2024_full.xlsx"

# Đọc sheet "Wide" (chứa dữ liệu dạng hàng-cột)
df = pd.read_excel(file_path, sheet_name="Wide")

# Chuyển dữ liệu sang dạng dọc (long format)
df_long = df.melt(id_vars=["Chỉ số"], var_name="Tháng", value_name="Giá trị")

# Lọc ra dữ liệu cho Thất nghiệp và PPI
df_selected = df_long[df_long["Chỉ số"].isin(["Thất nghiệp (%)", "PPI (Producer Price Index)"])]

# Chuyển ngược lại sang dạng rộng để hiển thị cột riêng
df_final = df_selected.pivot(index="Tháng", columns="Chỉ số", values="Giá trị").reset_index()

# Đổi tên cột
df_final = df_final.rename(columns={
    "Tháng": "Tháng",
    "Thất nghiệp (%)": "Thất nghiệp",
    "PPI (Producer Price Index)": "PPI"
})

# Hiển thị kết quả
print(df_final[["Tháng", "Thất nghiệp", "PPI"]])
```

Bài 1.9: File dữ liệu JSON **data/vn10_stocks.json** chứa thông tin về  chỉ số  chứng khoán của một số cổ phiếu. Hãy đọc file JSON này rồi chuyển thành DataFrame.

```python
import json

json_path = root_path + "data/vn10_stocks.json"

# --- Parse kiểu "trong ảnh" ---
# 1) Đọc toàn bộ file thành chuỗi
with open(json_path, "r", encoding="utf-8") as f:
    obj = f.read()

# 2) json.loads: chuỗi JSON -> đối tượng Python (dict/list)
result = json.loads(obj)

# 3) json.dumps: đối tượng Python -> chuỗi JSON
asjson = json.dumps(result, ensure_ascii=False)

# 4) Đưa về cấu trúc có khóa 'stocks' để dễ tạo DataFrame
if isinstance(result, list):
    result = {"stocks": result}
elif isinstance(result, dict) and "stocks" not in result:
    # nếu JSON bọc trong 1 khóa khác chứa list
    for k, v in result.items():
        if isinstance(v, list):
            result = {"stocks": v}
            break

# 5) Tạo DataFrame từ danh sách cổ phiếu, chọn vài cột tiêu biểu
df = pd.DataFrame(result["stocks"], columns=["ticker", "company", "price", "pe", "pb"])
print(df)
```

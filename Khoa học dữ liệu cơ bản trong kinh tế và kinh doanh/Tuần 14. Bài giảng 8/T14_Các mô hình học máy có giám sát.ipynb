## 1. Giới thiệu về Học máy

### Học máy là gì?

Học máy (Machine Learning - ML) là một lĩnh vực của trí tuệ nhân tạo (AI), tập trung vào việc sử dụng dữ liệu và thuật toán để bắt chước cách con người học, dần dần cải thiện độ chính xác của nó.

Nói một cách đơn giản, thay vì lập trình một cách tường minh các quy tắc để giải quyết vấn đề, chúng ta "dạy" cho máy tính học các quy tắc đó từ dữ liệu.



### Các loại hình học máy chính:

* **Học có giám sát (Supervised Learning):** Dữ liệu đầu vào đã được gán nhãn (dữ liệu đã được gán nhãn). Mô hình học cách ánh xạ từ đầu vào (các **đặc trưng - features**) đến đầu ra (các **nhãn - labels**). Đây là loại hình phổ biến nhất.
    * *Ví dụ:* Dự đoán giá nhà (đầu vào: diện tích, số phòng ngủ; đầu ra: giá nhà), phân loại email là spam hay không (đầu vào: nội dung email; đầu ra: nhãn "spam" / "không spam").
* **Học không giám sát (Unsupervised Learning):** Dữ liệu đầu vào không được gán nhãn. Mô hình tự tìm ra các cấu trúc, quy luật hoặc **cụm (cluster)** trong dữ liệu.
    * *Ví dụ:* Phân cụm khách hàng dựa trên hành vi mua sắm, giảm chiều dữ liệu.
* **Học củng cố (Reinforcement Learning):** Mô hình học thông qua phương pháp **thử và sai**. Nó tương tác với một môi trường và nhận **phần thưởng** hoặc **hình phạt** cho mỗi hành động.
    * *Ví dụ:* Huấn luyện bot chơi game, xe tự lái.

## 2. Các mô hình và Quy trình làm việc cơ bản

Trong bài giảng này, chúng ta sẽ tập trung vào **Học có giám sát** với hai dạng bài toán chính:

* **Bài toán Hồi quy (Regression):** Dự đoán một giá trị liên tục. Ví dụ: dự đoán nhiệt độ, giá cổ phiếu.
* **Bài toán Phân loại (Classification):** Dự đoán một nhãn/danh mục rời rạc. Ví dụ: phân loại chó/mèo, chẩn đoán bệnh.



### 2.1 Các mô hình sẽ học trong bài giảng

| Tên mô hình | Bài toán áp dụng | Mô tả ngắn |
| :--- | :--- | :--- |
| **Linear Regression** | Hồi quy | Tìm một đường thẳng phù hợp nhất với dữ liệu để dự đoán giá trị liên tục. |
| **Logistic Regression** | Phân loại | Vẫn dùng một đường (ranh giới) để chia dữ liệu, nhưng là để phân loại chúng vào các nhóm khác nhau dựa trên xác suất. |
| **K-Nearest Neighbors (KNN)** | Hồi quy & Phân loại | Dự đoán dựa trên "hàng xóm" gần nhất. Với phân loại, nó lấy theo số đông; với hồi quy, nó lấy giá trị trung bình. |
| **Decision Tree** | Hồi quy & Phân loại | Xây dựng một mô hình dạng cây gồm các quy tắc "nếu-thì" để đưa ra quyết định. |
| **Random Forest** | Hồi quy & Phân loại | Giống như một cuộc bỏ phiếu của nhiều Cây quyết định (một "khu rừng"). Kết quả cuối cùng được quyết định bởi số đông để tăng độ chính xác. |


### 2.2 Quy tắc chia Train-Test và các chỉ số đánh giá cơ bản

Trong học máy, chúng ta không bao giờ sử dụng toàn bộ dữ liệu để huấn luyện mô hình. Thay vào đó, chúng ta chia bộ dữ liệu thành hai phần:
- **Tập huấn luyện (Training Set):** Chiếm phần lớn dữ liệu (thường là 70-80%), được dùng để "dạy" cho mô hình.
- **Tập kiểm tra (Test Set):** Phần còn lại (20-30%), được dùng để đánh giá hiệu suất của mô hình sau khi đã huấn luyện xong. Tập test giống như "đề thi cuối kỳ" mà mô hình chưa từng thấy trước đó.

**Tại sao phải làm vậy?** Để kiểm tra khả năng **tổng quát hóa** của mô hình. Nếu một mô hình hoạt động tốt trên tập train nhưng kém trên tập test, nó đang bị **quá khớp (overfitting)** hay học vẹt.



#### Chỉ số đánh giá cho bài toán Hồi quy
**Mean Squared Error (MSE):** Trung bình của bình phương sai số. Nó "phạt" rất nặng những dự đoán sai lệch lớn.
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_{\text{thực tế}} - y_{\text{dự đoán}})^2$$

**Root Mean Squared Error (RMSE):** Là căn bậc hai của MSE, đưa đơn vị của sai số về cùng đơn vị với biến mục tiêu, giúp dễ diễn giải hơn.
$$RMSE = \sqrt{MSE}$$

#### Chỉ số đánh giá cho bài toán Phân loại
**Accuracy (Độ chính xác):** Tỷ lệ phần trăm số điểm được dự đoán đúng trên tổng số điểm.
$$Accuracy = \frac{\text{Số dự đoán đúng}}{\text{Tổng số dự đoán}}$$

### 2.3 Quá khớp và Dưới khớp

Sau khi hiểu tại sao cần chia dữ liệu để kiểm tra, chúng ta sẽ tìm hiểu kỹ hơn hai vấn đề phổ biến nhất mà việc này giúp phát hiện.


* **Dưới khớp (Underfitting):**
    * **Hiện tượng:** Mô hình quá đơn giản, không nắm bắt được quy luật cơ bản của dữ liệu. Nó hoạt động kém trên cả tập train và tập test.
    * **Ví dụ:** Dùng một đường thẳng để mô tả một mối quan hệ hình parabol.
    * **Cách khắc phục:** Thử mô hình phức tạp hơn, thêm đặc trưng cho dữ liệu.

* **Quá khớp (Overfitting):**
    * **Hiện tượng:** Mô hình quá phức tạp, nó học thuộc lòng cả **nhiễu (noise)** trong tập train. Nó hoạt động cực kỳ tốt trên tập train nhưng rất kém trên tập test.
    * **Ví dụ:** Dùng một đường cong uốn lượn đi qua chính xác tất cả các điểm dữ liệu trong tập train.
    * **Cách khắc phục:** Dùng nhiều dữ liệu hơn, giảm độ phức tạp của mô hình, sử dụng các kỹ thuật **điều chuẩn (regularization)**.

### 2.4 Quy trình chuẩn của một dự án Học máy
Một dự án học máy điển hình không chỉ có bước huấn luyện mô hình, mà là một quy trình gồm nhiều giai đoạn lặp đi lặp lại.
[Hình ảnh về quy trình một dự án học máy]
1.  **Thu thập dữ liệu (Data Collection):** Tìm kiếm và thu thập dữ liệu thô từ nhiều nguồn khác nhau.
2.  **Tiền xử lý & Phân tích dữ liệu khám phá (Data Preprocessing & EDA):** Giai đoạn quan trọng nhất. Bao gồm làm sạch dữ liệu (xử lý giá trị thiếu, **giá trị ngoại lai - outliers**), chuẩn hóa, và khám phá dữ liệu để tìm ra các thông tin hữu ích ban đầu.
3.  **Lựa chọn & Huấn luyện mô hình (Model Selection & Training):** Dựa trên bài toán và dữ liệu, chọn ra một vài mô hình tiềm năng và huấn luyện chúng trên tập huấn luyện.
4.  **Đánh giá mô hình (Model Evaluation):** Sử dụng tập kiểm tra để đánh giá hiệu suất của các mô hình, chọn ra mô hình tốt nhất.
5.  **Tinh chỉnh tham số (Parameter Tuning):** Tối ưu hóa các tham số của mô hình tốt nhất để cải thiện hiệu suất.
6.  **Triển khai (Deployment):** Đưa mô hình vào môi trường thực tế để nó có thể đưa ra dự đoán trên dữ liệu mới.

## 3. Cách mô hình học từ dữ liệu: Tính tay và Thực hành

Trong phần này, chúng ta sẽ đi sâu vào từng mô hình. Với mỗi mô hình, chúng ta sẽ tìm hiểu cách nó học thông qua ví dụ tính tay, ngay sau đó là phần thực hành với thư viện Scikit-learn để đối chiếu kết quả.

Để thực hành, chúng ta sẽ sử dụng **Scikit-learn**, thư viện học máy phổ biến nhất trong Python. Điểm mạnh của Scikit-learn là giao diện lập trình nhất quán và dễ sử dụng. Quy trình chung thường bao gồm 3 bước: 
+ `1. Khởi tạo mô hình`, 
+ `2. Huấn luyện (.fit())`, 
+ `3. Dự đoán (.predict())`.

```python
# Khởi tạo các thư viện và dữ liệu dùng chung cho Phần 3
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, roc_curve, roc_auc_score

# Dữ liệu Hồi quy
X_reg = np.array([[1], [1.5], [2.5], [3], [4], [5], [5.5], [6], [7], [8]])
y_reg = np.array([15, 18, 22, 25, 28, 35, 36, 40, 45, 52])
X_train_reg, X_test_reg = X_reg[:8], X_reg[8:]
y_train_reg, y_test_reg = y_reg[:8], y_reg[8:]

# Dữ liệu Phân loại
X_cls = np.array([[1], [2], [3], [4.5], [5.5], [6], [7], [8], [4], [9]])
y_cls = np.array([0, 0, 0, 0, 1, 1, 1, 1, 0, 1]) # 0: Trượt, 1: Đạt
X_train_cls, X_test_cls = X_cls[:8], X_cls[8:]
y_train_cls, y_test_cls = y_cls[:8], y_cls[8:]
```

### 3.1 Bài toán hồi quy

#### 3.1.1 Bài toán và dữ liệu

**Bài toán:** Dự đoán lương (triệu VND) dựa trên số năm kinh nghiệm.

**Dữ liệu (10 quan sát):**
| Kinh nghiệm (năm) | Lương (triệu) |
|:---:|:---:|
| 1 | 15 |
| 1.5 | 18 |
| 2.5 | 22 |
| 3 | 25 |
| 4 | 28 |
| 5 | 35 |
| 5.5 | 36 |
| 6 | 40 |
| 7 | 45 |
| 8 | 52 |

**Quy tắc chia:**
- **Tập Train:** 8 điểm dữ liệu đầu tiên.
- **Tập Test:** 2 điểm dữ liệu cuối `(7, 45)` và `(8, 52)`.

#### 3.1.2 Linear Regression
##### Mô hình học như thế nào?
Mô hình Hồi quy tuyến tính cố gắng tìm ra một phương trình đường thẳng `y_dự_đoán = β₀ + β₁ * x` phù hợp nhất với **dữ liệu tập train**. Trong đó:
- `β₁` là **hệ số góc (slope)**: Cho biết Lương thay đổi bao nhiêu khi Kinh nghiệm tăng 1 năm.
- `β₀` là **hệ số chặn (intercept)**: Là mức lương cơ bản khi kinh nghiệm bằng 0.
Nó "học" bằng cách điều chỉnh `β₀` và `β₁` liên tục sao cho tổng bình phương sai số (MSE) giữa giá trị dự đoán và giá trị thực tế trên **tập train** là nhỏ nhất.



Sau khi huấn luyện, mô hình học được các hệ số là **`β₀ ≈ 10.19`** và **`β₁ ≈ 4.82`**. Ta sẽ dùng các hệ số này để tính toán.
**Lương ≈ 10.19 + 4.82 * Kinh nghiệm**

**Dự đoán và Đánh giá:**
- Với Kinh nghiệm = 7: `y_dự_đoán` = 10.19 + 4.82 * 7 ≈ **43.95** (Thực tế: 45)
- Với Kinh nghiệm = 8: `y_dự_đoán` = 10.19 + 4.82 * 8 ≈ **48.78** (Thực tế: 52)

**Tính RMSE:**
$$RMSE = \sqrt{\frac{(45 - 43.95)^2 + (52 - 48.78)^2}{2}} \approx 2.395$$

##### Thực hành với Scikit-learn

```python
model_lr = LinearRegression()
model_lr.fit(X_train_reg, y_train_reg)
preds_lr = model_lr.predict(X_test_reg)
rmse_lr = np.sqrt(mean_squared_error(y_test_reg, preds_lr))
print(f"Hệ số β₁ (slope): {model_lr.coef_[0]:.4f}")
print(f"Hệ số β₀ (intercept): {model_lr.intercept_:.4f}")
print(f"Dự đoán: {preds_lr.flatten()}")
print(f"RMSE: {rmse_lr:.4f}")
```

```python
# Trực quan hóa
plt.figure(figsize=(10, 6))
plt.scatter(X_train_reg, y_train_reg, color='blue', label='Dữ liệu Train')
plt.scatter(X_test_reg, y_test_reg, color='green', label='Dữ liệu Test (Thực tế)')
plt.scatter(X_test_reg, preds_lr, color='red', marker='x', s=100, label='Dữ liệu Test (Dự đoán)')
# Vẽ đường hồi quy
X_plot = np.linspace(0, 10, 100).reshape(-1, 1)
y_plot = model_lr.predict(X_plot)
plt.plot(X_plot, y_plot, color='red', linestyle='--', label='Đường Hồi quy Tuyến tính')

plt.title('Linear Regression - Trực quan hóa')
plt.xlabel('Kinh nghiệm (năm)')
plt.ylabel('Lương (triệu)')
plt.legend()
plt.grid(True)
plt.savefig('ml_image1.png')
```

#### 3.1.3 K-Nearest Neighbors (KNN)
##### Mô hình học như thế nào?
KNN là một **"mô hình học lười" (lazy learner)**. Nó không thực sự học một hàm số nào cả. Thay vào đó, nó ghi nhớ toàn bộ **dữ liệu tập train**. Khi cần dự đoán cho một điểm dữ liệu mới (từ tập test), nó sẽ:
1. Tính khoảng cách từ điểm mới này đến tất cả các điểm trong tập train.
2. Tìm ra K điểm trong tập train có khoảng cách gần nhất (K "hàng xóm" gần nhất).
3. Lấy trung bình giá trị (lương) của K hàng xóm đó để làm kết quả dự đoán.



Xét **K=3**.

**Dự đoán và Đánh giá:**
- Với Kinh nghiệm = 7 (hàng xóm: 6, 5.5, 5): `y_dự_đoán` = (40 + 36 + 35) / 3 = **37** (Thực tế: 45)
- Với Kinh nghiệm = 8 (hàng xóm: 6, 5.5, 5): `y_dự_đoán` = (40 + 36 + 35) / 3 = **37** (Thực tế: 52)

**Tính RMSE:**
$$RMSE = \sqrt{\frac{(45 - 37)^2 + (52 - 37)^2}{2}} \approx 12.02$$

##### Thực hành với Scikit-learn

```python
model_knn_reg = KNeighborsRegressor(n_neighbors=3)
model_knn_reg.fit(X_train_reg, y_train_reg)
preds_knn_reg = model_knn_reg.predict(X_test_reg)
rmse_knn_reg = np.sqrt(mean_squared_error(y_test_reg, preds_knn_reg))
print(f"Dự đoán: {preds_knn_reg.flatten()}")
print(f"RMSE: {rmse_knn_reg:.4f}")
```

```python
# Trực quan hóa
plt.figure(figsize=(10, 6))
plt.scatter(X_train_reg, y_train_reg, color='blue', label='Dữ liệu Train')
plt.scatter(X_test_reg, y_test_reg, color='green', label='Dữ liệu Test (Thực tế)')
plt.scatter(X_test_reg, preds_knn_reg, color='red', marker='x', s=100, label='Dữ liệu Test (Dự đoán)')

# Vẽ đường dự đoán của KNN
X_plot = np.linspace(0, 10, 100).reshape(-1, 1)
y_plot = model_knn_reg.predict(X_plot)
plt.plot(X_plot, y_plot, color='red', linestyle='--', label='Đường dự đoán KNN (k=3)')

plt.title('KNN Regressor - Trực quan hóa')
plt.xlabel('Kinh nghiệm (năm)')
plt.ylabel('Lương (triệu)')
plt.legend()
plt.grid(True)
plt.savefig('ml_image2.png')
```

#### 3.1.4 Decision Tree
##### Mô hình học như thế nào?
Hãy tưởng tượng mô hình Cây quyết định học bằng cách đặt ra một loạt câu hỏi "Có/Không" để chia **dữ liệu tập train**. Ví dụ: `Kinh nghiệm < 3.5 năm?`. Tại mỗi bước, nó sẽ tìm ra câu hỏi (dựa trên một đặc trưng và một ngưỡng) giúp phân chia dữ liệu thành hai nhóm con. Tiêu chí để chọn câu hỏi tốt nhất là tìm phép chia sao cho **trung bình bình phương sai số (MSE)** của các nhóm con là nhỏ nhất. Điều này đảm bảo các giá trị trong mỗi nhóm con là gần nhau nhất có thể. Quá trình này lặp lại cho đến khi đạt được một điều kiện dừng. Kết quả là một cây quy tắc. Khi dự đoán, một điểm dữ liệu mới sẽ đi từ gốc xuống các nhánh theo các quy tắc và nhận giá trị dự đoán là trung bình của các điểm trong lá cuối cùng mà nó rơi vào.


**Tính toán ban đầu (nút gốc):**
Trước khi chia, mô hình cần một điểm tham chiếu. Nó sẽ tính MSE cho toàn bộ tập huấn luyện. Đây là mức sai số ban đầu nếu chúng ta chỉ dự đoán bằng giá trị trung bình của toàn bộ tập dữ liệu.
- Trung bình của lương trong tập train: `(15+18+22+25+28+35+36+40) / 8 = 27.375`
- MSE ban đầu: `((15-27.375)² + ... + (40-27.375)²) / 8 ≈ 71.23`
Mục tiêu của việc tìm ngưỡng chia là để giảm chỉ số này xuống thấp nhất có thể.

**Tìm ngưỡng chia tối ưu (cho nút gốc):**
Mô hình sẽ thử tất cả các ngưỡng chia khả dĩ (điểm giữa của hai giá trị kinh nghiệm liên tiếp) và chọn ngưỡng nào cho **MSE sau chia** thấp nhất. Công thức tính là: $$MSE_{\text{sau chia}} = \frac{n_{\text{trái}}}{n_{\text{tổng}}} MSE_{\text{trái}} + \frac{n_{\text{phải}}}{n_{\text{tổng}}} MSE_{\text{phải}}$$

| Ngưỡng chia (<) | MSE (Trái) | MSE (Phải) | MSE sau chia |
|:---:|:---:|:---:|:---:|
| 1.25 | 0.0 | 55.27 | 48.36 |
| 2.0 | 2.25 | 33.56 | 25.73 |
| 2.75 | 8.22 | 26.24 | 19.48 |
| 3.5 | 14.5 | 18.69 | 16.60 |
| **4.5** | **21.84** | **4.67** | **15.40** |
| 5.25 | 41.81 | 4.0 | 32.35 |
| 5.75 | 47.63 | 0.0 | 41.68 |

-> Mô hình chọn `Kinh nghiệm < 4.5` là quy tắc chia đầu tiên vì nó cho MSE sau chia thấp nhất (giảm từ 71.23 xuống 15.40). Giá trị dự đoán cho nhánh `Kinh nghiệm >= 4.5` sẽ là giá trị trung bình của các điểm trong nhánh đó: (35+36+40)/3 = 37.



**Dự đoán và Đánh giá:**
- Với Kinh nghiệm = 7 (>= 4.5): `y_dự_đoán` = **37** (Thực tế: 45)
- Với Kinh nghiệm = 8 (>= 4.5): `y_dự_đoán` = **37** (Thực tế: 52)

**Tính RMSE:**
$$RMSE = \sqrt{\frac{(45 - 37)^2 + (52 - 37)^2}{2}} \approx 12.02$$

#### 3.2.2 Logistic Regression
##### Mô hình học như thế nào?
Mô hình Hồi quy Logistic học cách biến đổi một giá trị đầu vào `x` thành một xác suất (từ 0 đến 1) bằng **hàm Sigmoid**. Xác suất này chính là `y_dự_đoán` cho lớp 1 ('Đạt').

Sau khi huấn luyện trên tập train, mô hình học được các hệ số tối ưu là **`β₀ ≈ -5.5288`** và **`β₁ ≈ 1.1590`**. Vậy công thức dự đoán xác suất là:
$$y_{\text{dự đoán}} = P(y=1|x) = \frac{1}{1 + e^{-(-5.5288 + 1.1590 * x)}}$$

Nếu xác suất này > 0.5, mô hình sẽ phân loại là 'Đạt', nếu không thì là 'Trượt'.



**Dự đoán và Đánh giá:**
- Với Số giờ = 4: $$P(y=1|x=4) = \frac{1}{1 + e^{-(-5.5288 + 1.1590 * 4)}} = \frac{1}{1 + e^{0.8928}} \approx 0.29$$ Vì 0.29 < 0.5, dự đoán là **Trượt**. (Đúng)
- Với Số giờ = 9: $$P(y=1|x=9) = \frac{1}{1 + e^{-(-5.5288 + 1.1590 * 9)}} = \frac{1}{1 + e^{-4.9022}} \approx 0.99$$ Vì 0.99 > 0.5, dự đoán là **Đạt**. (Đúng)

**Tính Accuracy:**
$$Accuracy = \frac{2 \text{ (dự đoán đúng)}}{2 \text{ (tổng số)}} = 100\%$$

##### Thực hành với Scikit-learn

```python
model_log_reg = LogisticRegression(random_state=42)
model_log_reg.fit(X_train_cls, y_train_cls)
preds_log_reg = model_log_reg.predict(X_test_cls)
acc_log_reg = accuracy_score(y_test_cls, preds_log_reg)
print(f"Hệ số β₁ (coef): {model_log_reg.coef_[0][0]:.4f}")
print(f"Hệ số β₀ (intercept): {model_log_reg.intercept_[0]:.4f}")
print(f"Dự đoán: {preds_log_reg}")
print(f"Accuracy: {acc_log_reg:.2f}")
```

#### 3.2.4 Decision Tree
##### Mô hình học như thế nào?
Từ **dữ liệu tập train**, mô hình học bằng cách liên tục đặt ra các câu hỏi "Có/Không" để chia dữ liệu. Nó sẽ duyệt qua tất cả các đặc trưng và các ngưỡng chia khả dĩ để tìm ra câu hỏi tốt nhất. "Tốt nhất" ở đây có nghĩa là câu hỏi đó tạo ra hai nhóm con "thuần khiết" nhất. Sự thuần khiết được đo bằng **chỉ số Gini**. Gini càng gần 0, nhóm càng thuần khiết. Quá trình này lặp lại, tạo ra một cấu trúc cây. Khi dự đoán, một điểm dữ liệu mới sẽ đi theo các câu hỏi trên cây để đến một nút lá, và được gán nhãn của lớp chiếm đa số trong nút lá đó.

**Công thức Gini:** Với C là số lớp, $p_i$ là tỷ lệ của lớp i trong một nút.
$$Gini = 1 - \sum_{i=1}^{C} p_i^2$$



**Tính toán ban đầu (nút gốc):**
Tương tự bài toán hồi quy, mô hình cũng cần một điểm tham chiếu. Nó sẽ tính chỉ số Gini cho toàn bộ tập huấn luyện để đo "độ hỗn loạn" ban đầu.
- Trong 8 điểm của tập train, có 4 điểm 'Trượt' (lớp 0) và 4 điểm 'Đạt' (lớp 1).
- Tỷ lệ lớp 0: p₀ = 4/8 = 0.5
- Tỷ lệ lớp 1: p₁ = 4/8 = 0.5
- Gini ban đầu: $$Gini = 1 - (0.5^2 + 0.5^2) = 1 - (0.25 + 0.25) = 0.5$$
Mục tiêu của việc tìm ngưỡng chia là để giảm chỉ số Gini này xuống thấp nhất có thể.

**Tìm ngưỡng chia tối ưu (cho nút gốc):**
Mô hình sẽ thử các ngưỡng chia và chọn ngưỡng nào có **Gini sau chia** thấp nhất. Công thức tính là: $$Gini_{\text{sau chia}} = \frac{n_{\text{trái}}}{n_{\text{tổng}}}Gini_{\text{trái}} + \frac{n_{\text{phải}}}{n_{\text{tổng}}}Gini_{\text{phải}}$$

| Ngưỡng chia (<) | Gini (Trái) | Gini (Phải) | Gini sau chia |
|:---:|:---:|:---:|:---:|
| 1.5 | 0.0 | 0.490 | 0.428 |
| 2.5 | 0.0 | 0.444 | 0.333 |
| 3.75 | 0.0 | 0.320 | 0.200 |
| **5.0** | **0.0** | **0.0** | **0.0** |
| 5.75 | 0.320 | 0.0 | 0.200 |
| 6.5 | 0.444 | 0.0 | 0.333 |
| 7.5 | 0.490 | 0.0 | 0.428 |

-> Mô hình chọn `Giờ học < 5.0` là quy tắc chia tối ưu vì nó cho Gini sau chia thấp nhất (giảm từ 0.5 xuống 0.0), tạo ra hai nhóm con hoàn toàn thuần khiết.


**Dự đoán và Đánh giá:**
- Với Số giờ = 4 (< 5.0): `y_dự_đoán` = **Trượt**. (Đúng)
- Với Số giờ = 9 (>= 5.0): `y_dự_đoán` = **Đạt**. (Đúng)

**Tính Accuracy:**
$$Accuracy = \frac{2}{2} = 100\%$$

##### Thực hành với Scikit-learn

```python
# Huấn luyện mô hình với max_depth=1 để khớp với tính toán tay
model_dt_cls = DecisionTreeClassifier(max_depth=1, random_state=42)
model_dt_cls.fit(X_train_cls, y_train_cls)
preds_dt_cls = model_dt_cls.predict(X_test_cls)
acc_dt_cls = accuracy_score(y_test_cls, preds_dt_cls)
print(f"Dự đoán: {preds_dt_cls}")
print(f"Accuracy: {acc_dt_cls:.2f}")
```

#### 3.2.5 Random Forest
##### Mô hình học như thế nào?
Tương tự như Random Forest hồi quy, mô hình xây dựng một "khu rừng" gồm nhiều Cây quyết định khác nhau từ **tập train**. Khi dự đoán, mỗi cây trong rừng sẽ đưa ra một "phiếu bầu" cho một lớp. Lớp nào nhận được nhiều phiếu bầu nhất sẽ là kết quả dự đoán cuối cùng. Cách tiếp cận "trí tuệ tập thể" này giúp mô hình trở nên ổn định và chính xác hơn một cây đơn lẻ.

### 3.3 Bàn luận thêm
#### Khi nào các thuật toán dừng học?
Quá trình "học" không diễn ra mãi mãi. Các mô hình có những điều kiện dừng khác nhau:
* **Các mô hình lặp (Linear/Logistic Regression):** Thường dừng sau một số lượng vòng lặp (`max_iter`) được định sẵn, hoặc khi mức độ cải thiện (ví dụ: sự thay đổi của hàm mất mát) giữa các vòng lặp trở nên rất nhỏ, không đáng kể.
* **Cây quyết định (Decision Tree):** Dừng khi một nhánh (nút) không thể chia nhỏ hơn được nữa. Điều này xảy ra khi:
    * Tất cả các điểm dữ liệu trong nút đó đều thuộc cùng một lớp (nút "thuần khiết").
    * Đã đạt đến độ sâu tối đa của cây (`max_depth`) mà người dùng cho phép.
    * Số lượng mẫu trong một nút quá nhỏ để có thể chia tiếp (`min_samples_split`).
* **KNN:** Là "mô hình học lười", nó không có quá trình huấn luyện thực sự nên không có điều kiện dừng học. Nó chỉ đơn giản là lưu trữ toàn bộ dữ liệu.



#### Khi nào nên Chuẩn hóa dữ liệu (Data Scaling)?
**Chuẩn hóa dữ liệu** là quá trình đưa các cột dữ liệu (đặc trưng) về cùng một **thang đo (scale)**. Đây là một bước cực kỳ quan trọng.

* **Tại sao cần thiết?** Hãy tưởng tượng bạn có 2 đặc trưng: `kinh nghiệm` (từ 1-10) và `lương tháng trước` (từ 15,000,000 - 50,000,000). Với các thuật toán dựa trên **khoảng cách** như **KNN**, sự chênh lệch tuyệt đối về giá trị của đặc trưng `lương` sẽ hoàn toàn "lấn át" đặc trưng `kinh nghiệm`. Mô hình sẽ gần như chỉ dựa vào lương để tính khoảng cách, bỏ qua thông tin quan trọng từ kinh nghiệm.

* **Mô hình nào cần?**
    * **CẦN:** Các mô hình nhạy cảm với thang đo của dữ liệu, bao gồm các mô hình dựa trên khoảng cách (**KNN**) và các mô hình tối ưu bằng **Hạ Gradient (Gradient Descent)** như **Linear Regression, Logistic Regression**.
    * **KHÔNG CẦN:** Các mô hình dựa trên quy tắc như **Decision Tree** và **Random Forest**. Các mô hình này xem xét từng đặc trưng một cách độc lập tại mỗi lần chia, nên không bị ảnh hưởng bởi sự khác biệt về thang đo giữa các đặc trưng.

## 4. Các chỉ số đánh giá nâng cao

Các chỉ số cơ bản như RMSE và Độ chính xác (Accuracy) rất hữu ích, nhưng chúng không phải lúc nào cũng kể toàn bộ câu chuyện.

### 4.1 Với bài toán hồi quy
**Lỗi tuyệt đối trung bình (Mean Absolute Error - MAE):** Trung bình của trị tuyệt đối sai số. Chỉ số này ít bị ảnh hưởng bởi các giá trị ngoại lai hơn MSE.
$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{\text{thực tế}} - y_{\text{dự đoán}}|$$

**Hệ số xác định R² (R-squared):** Cho chúng ta biết mô hình tốt đến mức nào so với một mô hình đơn giản (ví dụ: luôn dự đoán giá trị trung bình). Nó cho biết bao nhiêu phần trăm sự biến thiên của dữ liệu được mô hình giải thích. R² càng gần 1, mô hình càng tốt.
$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_{\text{thực tế}, i} - y_{\text{dự đoán}, i})^2}{\sum_{i=1}^{n} (y_{\text{thực tế}, i} - \bar{y}_{\text{thực tế}})^2}$$
Trong đó $\bar{y}_{\text{thực tế}}$ là giá trị trung bình của tất cả các điểm dữ liệu thực tế.



### 4.2 Với bài toán phân loại
**Vấn đề của Độ chính xác (Accuracy):** Hãy tưởng tượng một bài toán dự đoán bệnh hiếm gặp, trong đó 99% người là "bình thường" và 1% là "bệnh". Một mô hình ngây thơ luôn dự đoán "bình thường" sẽ có độ chính xác là 99%! Tuy nhiên, mô hình này hoàn toàn vô dụng vì nó không phát hiện được bất kỳ trường hợp bệnh nào.

Đây là lúc chúng ta cần các chỉ số khác:

* **Ma trận nhầm lẫn (Confusion Matrix):** Một bảng cho thấy chi tiết các dự đoán đúng và sai.
    * **Dương tính thật (True Positive - TP):** Dự đoán 'Bệnh' và thực tế là 'Bệnh'. (Tốt)
    * **Âm tính thật (True Negative - TN):** Dự đoán 'Bình thường' và thực tế là 'Bình thường'. (Tốt)
    * **Dương tính giả (False Positive - FP):** Dự đoán 'Bệnh' nhưng thực tế là 'Bình thường'. (Báo động giả)
    * **Âm tính giả (False Negative - FN):** Dự đoán 'Bình thường' nhưng thực tế là 'Bệnh'. (Bỏ sót, rất nguy hiểm!)



* **Độ chuẩn (Precision):** Trong số tất cả các lần dự đoán là 'Bệnh', có bao nhiêu là đúng? Giúp giảm báo động giả.
$$Precision = \frac{TP}{TP + FP}$$

* **Độ phủ (Recall / Độ nhạy):** Trong số tất cả những người thực sự 'Bệnh', mô hình phát hiện được bao nhiêu? Giúp giảm bỏ sót.
$$Recall = \frac{TP}{TP + FN}$$

* **Điểm F1 (F1-Score):** Trung bình điều hòa của Độ chuẩn và Độ phủ, hữu ích khi cần cân bằng cả hai.

* **Đường cong ROC và chỉ số AUC:**
    * **Đường cong ROC (Receiver Operating Characteristic):** Là một biểu đồ thể hiện khả năng phân loại của một mô hình tại tất cả các **ngưỡng (thresholds)** khác nhau. Nó vẽ **Tỷ lệ Dương tính thật (True Positive Rate hay Recall)** theo **Tỷ lệ Dương tính giả (False Positive Rate)**.
    * **Diện tích dưới đường cong (Area Under the Curve - AUC):** Là diện tích phía dưới đường cong ROC. AUC đo lường toàn bộ hiệu suất hai chiều của mô hình. AUC càng gần 1, mô hình càng tốt. Một mô hình đoán ngẫu nhiên sẽ có AUC = 0.5.


```python
# Minh họa ROC và AUC cho mô hình Logistic Regression

# Lấy xác suất dự đoán cho lớp 1 ('Đạt') trên tập test
y_pred_proba = model_log_reg.predict_proba(X_test_cls)[:, 1]

# Tính toán các giá trị cho đường cong ROC
fpr, tpr, thresholds = roc_curve(y_test_cls, y_pred_proba)

# Tính chỉ số AUC
auc = roc_auc_score(y_test_cls, y_pred_proba)
print(f"Chỉ số AUC: {auc:.4f}")

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='orange', label=f'Đường cong ROC (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--', label='Đường đoán ngẫu nhiên')
plt.xlabel('Tỷ lệ Dương tính giả (False Positive Rate)')
plt.ylabel('Tỷ lệ Dương tính thật (True Positive Rate)')
plt.title('Đường cong ROC cho mô hình Logistic Regression')
plt.legend()
plt.grid(True)
plt.savefig('ml_image15.png')
```
